```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X_train = np.linspace(0, 10, 30)
y_train = 3 * X_train + 4 + np.random.randn(30) * 2
```

```python
def gradient_descent(X, y, lr=0.01, epochs=1000):
    m, b = 0, 0
    n = len(X)
    for _ in range(epochs):
        y_pred = m * X + b
        dm = (-2/n) * sum(X * (y - y_pred))
        db = (-2/n) * sum(y - y_pred)
        m -= lr * dm
        b -= lr * db
    return m, b

m, b = gradient_descent(X_train, y_train)
print(f"Trained Model: y = {m:.2f}x + {b:.2f}")
```

```python
X_test = np.array([2, 4, 6, 8, 10])
y_test = 3 * X_test + 4 + np.random.randn(5) * 2
```

```python
y_pred = m * X_test + b
```

```python
mse = np.mean((y_test - y_pred)**2)
```

```python
ss_total = np.sum((y_test - np.mean(y_test))**2)
ss_res = np.sum((y_test - y_pred)**2)
r2 = 1 - (ss_res / ss_total)
print(f"MSE: {mse:.2f}")
print(f"R² Score: {r2*100:.2f}%")
```

```python
plt.scatter(X_test, y_test, color='blue', label='Actual')
plt.plot(X_test, y_pred, color='red', label='Predicted Line')
plt.legend()
plt.show()
```
